```python
import numpy as np
```

```python
meshFileName='oresund.mesh'
objFileName=meshFileName[:-4]+"obj"

fast=False 

with open(meshFileName, "r") as a_file:
    nVerticies=0
    nElements=0
    i=0
    text=''
    for line in a_file:
        if i==0: 
            nVerticies=int(line.split()[2])
            i+=1
            continue
        if i<=nVerticies:
            text=text+'v '+line.split()[1]+' '+line.split()[2]+' '+line.split()[3]+'\n'
            i+=1
            continue
        if i==nVerticies+1:
            nElements=int(line.split()[0])
            i+=1
            continue
        myLine=line.split()
        text=text+'f '
        for k in range(1,len(myLine)): #-1):
            text=text+myLine[k]+' '
            #if k==len(myLine)-2 and not fast:
            #    # areas without BC code definition assuming max 12 BC codes
            #    if int(myLine[k+1])>12: 
            #        text=text+myLine[k+1]+' '
        text=text+'\n'
        i+=1
    
with open(objFileName, 'w') as f:
    f.write(text)
```
