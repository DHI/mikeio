## Directional spectra

```python
fn = "../tests/testdata/line_dir_spectra.dfsu"
dfs = Dfsu(fn)
dfs
```

```python
ds = dfs.read()
spec = ds[0][1,2]
```

```python
plt.plot(dfs.directions*180/np.pi, spec)
plt.xlabel("directions [degrees]")
plt.ylabel("directional spectral energy [m*m*s]");
plt.title("Line dfsu file, directional spectrum in element 3");
```
