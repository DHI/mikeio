# Grid3D

``` python
Grid3D(
    self,
    *,
    x=None,
    x0=0.0,
    dx=None,
    nx=None,
    y=None,
    y0=0.0,
    dy=None,
    ny=None,
    z=None,
    z0=0.0,
    dz=None,
    nz=None,
    projection='NON-UTM',
    origin=(0.0, 0.0),
    orientation=0.0,
)
```

3D grid.

Origin in the center of cell in lower-left corner x, y and z axes are
increasing and equidistant

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [dx](#mikeio.Grid3D.dx) | X-axis grid spacing. |
| [dy](#mikeio.Grid3D.dy) | Y-axis grid spacing. |
| [dz](#mikeio.Grid3D.dz) | Z-axis grid spacing. |
| [is_geo](#mikeio.Grid3D.is_geo) | Are coordinates geographical (LONG/LAT)? |
| [is_local_coordinates](#mikeio.Grid3D.is_local_coordinates) | Are coordinates relative (NON-UTM)? |
| [nx](#mikeio.Grid3D.nx) | Number of x grid points. |
| [ny](#mikeio.Grid3D.ny) | Number of y grid points. |
| [nz](#mikeio.Grid3D.nz) | Number of z grid points. |
| [orientation](#mikeio.Grid3D.orientation) | Grid orientation. |
| [origin](#mikeio.Grid3D.origin) | Coordinates of grid origo (in projection). |
| [projection](#mikeio.Grid3D.projection) | The projection. |
| [projection_string](#mikeio.Grid3D.projection_string) | The projection string. |
| [x](#mikeio.Grid3D.x) | Array of x-axis coordinates (element center). |
| [y](#mikeio.Grid3D.y) | Array of y-axis coordinates (element center). |
| [z](#mikeio.Grid3D.z) | Array of z-axis node coordinates. |

## Methods

| Name                        | Description                               |
|-----------------------------|-------------------------------------------|
| [isel](#mikeio.Grid3D.isel) | Get a subset geometry from this geometry. |

### isel

``` python
Grid3D.isel(idx, axis)
```
