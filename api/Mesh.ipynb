# Mesh

``` python
Mesh(self, filename)
```

The Mesh class is initialized with a mesh file.

## Parameters

| Name     | Type        | Description   | Default    |
|----------|-------------|---------------|------------|
| filename | str \| Path | mesh filename | *required* |

## Attributes

| Name     | Type         | Description            |
|----------|--------------|------------------------|
| geometry | GeometryFM2D | Flexible Mesh geometry |

## Examples

````python
import mikeio
mikeio.Mesh("../data/odense_rough.mesh")
````

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [to_shapely](#mikeio.Mesh.to_shapely) | Convert Mesh geometry to shapely MultiPolygon. |
| [write](#mikeio.Mesh.write) | write mesh to file. |

### to_shapely

``` python
Mesh.to_shapely()
```

Convert Mesh geometry to shapely MultiPolygon.

#### Returns

| Name | Type         | Description                  |
|------|--------------|------------------------------|
|      | MultiPolygon | mesh as shapely MultiPolygon |

#### Examples

````python
import mikeio
msh = mikeio.open("../data/odense_rough.mesh")
msh.to_shapely()
````

### write

``` python
Mesh.write(outfilename)
```

write mesh to file.

#### Parameters

| Name        | Type | Description  | Default    |
|-------------|------|--------------|------------|
| outfilename | str  | path to file | *required* |
