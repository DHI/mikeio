# dataset.\_data_plot.\_DataArrayPlotterFM

``` python
dataset._data_plot._DataArrayPlotterFM(self, da)
```

Plot a DataArray with a GeometryFM geometry.

If DataArray has multiple time steps, the first step will be plotted.

If DataArray is 3D the surface layer will be plotted.

## Examples

````python
import mikeio
da = mikeio.read("../data/HD2D.dfsu")["Surface elevation"]
da.plot()
````

| [hist](#mikeio.dataset._data_plot._DataArrayPlotterFM.hist) | Plot DataArray as histogram (using ax.hist). |
